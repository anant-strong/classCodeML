1. Read a CSV file with data in multiple columns and compute mean, median mode of each of the columns.[link text](https://)

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
df = pd.read_csv('Housing.csv')
df['price']=df['price']/1000
# print(df['price'])
price = df['price']

area = df['area']
bedrooms = df['bedrooms']
stories = df['stories']
parking = df['parking']

print("Mean of price,area,bedrooms,stories,parking")
print(price.mean())
print(area.mean())
print(bedrooms.mean())
print(stories.mean())
print(parking.mean())

print("Median of price,area,bedrooms,stories,parking")
print(price.median())
print(area.median())
print(bedrooms.median())
print(stories.median())
print(parking.median())

print("Mode of price,area,bedrooms,stories,parking")
print(price.mode())
print(area.mode())
print(bedrooms.mode())
print(stories.mode())
print(parking.mode())
```

3. Create a game (like lottery/ rock paper/scissors) and write a python code to record how many times the computer won and the user won in an excel file. Plot the graph showing the success rate of user and computer with different colors


```python
import random
import pandas as pd
import matplotlib.pyplot as plt

# Function to play the game
def play_game():
    options = ['rock', 'paper', 'scissors']
    user_wins = 0
    computer_wins = 0
    total_games = 0

    while True:
        user_choice = input("Enter your choice (rock, paper, scissors) or 'q' to quit: ")
        if user_choice == 'q':
            break

        computer_choice = random.choice(options)

        if user_choice not in options:
            print("Invalid choice. Please try again.")
            continue

        print("User choice:", user_choice)
        print("Computer choice:", computer_choice)

        if user_choice == computer_choice:
            print("It's a tie!")
        elif (
            (user_choice == 'rock' and computer_choice == 'scissors')
            or (user_choice == 'paper' and computer_choice == 'rock')
            or (user_choice == 'scissors' and computer_choice == 'paper')
        ):
            print("You win!")
            user_wins += 1
        else:
            print("Computer wins!")
            computer_wins += 1

        total_games += 1
        print("")

    return user_wins, computer_wins, total_games

# Play the game
user_wins, computer_wins, total_games = play_game()

# Create a DataFrame to store the results
data = {'User Wins': [user_wins], 'Computer Wins': [computer_wins]}
df = pd.DataFrame(data)

# Write the results to an Excel file
df.to_excel('game_results.xlsx', index=False)

# Plot the success rate graph
success_rate = [user_wins/total_games, computer_wins/total_games]
colors = ['green', 'red']
labels = ['User', 'Computer']

plt.bar(labels, success_rate, color=colors)
plt.xlabel('Player')
plt.ylabel('Success Rate')
plt.title('Success Rate of User vs Computer')
plt.show()
```
